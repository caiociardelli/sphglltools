####Gaussian function

The Gaussian function possesses numerous applications and appears throughout *physics*, *mathematics*, *engineering*, *computer science*, and many other areas of knowledge.
<br>
<br>
It is defined as follows:
<br>
<br>
\begin{equation}
  g\,(x)
  \,=\,e^{-x^2}
\end{equation}
<br>
\begin{equation}
  \int_{-\infty}^{\infty} g\,(x)\,dx
  \,=\,\int_{-\infty}^{\infty} e^{-x^2}\,dx
  \,=\,\sqrt{\pi}
\end{equation}
<br>
<br>
\begin{equation}
  g_{\mu,\,\sigma}\,(x)
  \,=\,e^{-\frac{(x\,-\,\mu)^2}{2\sigma^2}}
\end{equation}
<br>
\begin{equation}
  \int_{-\infty}^{\infty} g_{\mu,\,\sigma}\,(x)\,dx
  \,=\,\int_{-\infty}^{\infty} e^{-\frac{(x\,-\,\mu)^2}{2\sigma^2}}\,dx
  \,=\,\sigma\sqrt{2\pi}
\end{equation}
<br>
<br>
It is usual to present the Gaussian in its normalized form, which means that its integral from $-\infty$ to $\infty$ is $1$:
<br>
<br>
\begin{equation}
  G_{\mu,\,\sigma}\,(x)
  \,=\,\frac{e^{-\frac{(x\,-\,\mu)^2}{2\sigma^2}}}{\sigma\sqrt{2\pi}}
\end{equation}
<br>
\begin{equation}
  \int_{-\infty}^{\infty} G_{\mu,\,\sigma}\,(x)\,dx
  \,=\,\frac{1}{\sigma\sqrt{2\pi}}\int_{-\infty}^{\infty} g_{\mu,\,\sigma}\,(x)\,dx
  \,=\,1
\end{equation}
<br>
<br>

####Probability density

The Gaussian function is commonly used in statistics to represent probability density. Let $\sigma$ be the standard deviation, corresponding to the width at half-height. In that case, we have:
<br>
<br>
\begin{equation}
  \int_{-\sigma}^{\sigma} G_{\sigma}\,(x)\,dx
  \,=\,\frac{1}{\sigma\sqrt{2\pi}}\int_{-\sigma}^{\sigma} e^{-\frac{x^2}{2\sigma^2}}\,dx
  \,\approx\,0.6827...
\end{equation}
<br>
<br>
\begin{equation}
  \int_{-2\sigma}^{2\sigma} G_{\sigma}\,(x)\,dx
  \,=\,\frac{1}{\sigma\sqrt{2\pi}}\int_{-2\sigma}^{2\sigma} e^{-\frac{x^2}{2\sigma^2}}\,dx
  \,\approx\,0.9545...
\end{equation}
<br>
<br>
\begin{equation}
  \int_{-3\sigma}^{3\sigma} G_{\sigma}\,(x)\,dx
  \,=\,\frac{1}{\sigma\sqrt{2\pi}}\int_{-3\sigma}^{3\sigma} e^{-\frac{x^2}{2\sigma^2}}\,dx
  \,\approx\,0.9973...
\end{equation}

####Filtering

The Fourier transform of a Gaussian is another Gaussian in the frequency domain. That property is useful for designing Gaussian filters. A common way of low-passing an arbitrary function $f\,(x)$ is convolving it with a Gaussian *kernel*. The larger the value of $\sigma$, the stronger the filter.
<br>
<br>
That procedure is known as *Gaussian smoothing*.
<br>
<br>

#####Dirac delta function

\begin{equation}
\delta\,(x)
\,=\,
  \begin{cases}
    \infty & \text{if}\quad x\,=\,0 \\
    0 & \text{otherwise}
  \end{cases}
\end{equation}
<br>
\begin{equation}
  \int_{-\infty}^{\infty} \delta\,(x)\,dx
  \,=\,1
\end{equation}
<br>
<br>
Shifting property
<br>
<br>
\begin{equation}
 \int_{-\infty}^{\infty} f\,(x)\,\delta\,\left(x\,-\,T\right)\,dx
 \,=\,f\,(T)
\end{equation}
<br>
<br>
Convolution
<br>
<br>
\begin{split}
  \left(f\,(x)\,*\,\delta\,\left(x\,-\,T\right)\right)
  \,&=\,
  \int_{-\infty}^{\infty} f\,(\tau)\,\delta\,\left(x\,-\,T\,-\,\tau\right)\,d\tau\\
  \,&=\,\int_{-\infty}^{\infty} f\,(\tau)\,\delta\,\left(\tau\,-\,\left(x\,-\,T\right)\right)\,d\tau\quad (\text{using that}\,\,\delta\,(x)\,=\,\delta\,(-x))\\
  \,&=\,f\,(x\,-\,T)
\end{split}
<br>
<br>
Gaussian filtering
<br>
<br>
\begin{equation}
  \left(G_{\sigma}\,(x)\,*\,\delta\,\left(x\,-\,T\right)\right)
  \,=\,\int_{-\infty}^{\infty} G_{\sigma}\,(\tau)\,\delta\,\left(x\,-\,T\,-\,\tau\right)\,d\tau
  \,=\,G_{\sigma}\,(x\,-\,T)
\end{equation}
<br>
<br>

```
def convolve (y, weights):
  """
  Convolves the y array with
  the Gaussian weights
  """
  m  = weights.size
  n  = y.size
  hw = m // 2

  c  = y.copy ()
  y *= 0

  for i in range (n):

    for j in range (m):

      k = i + j - hw
    
      if k >= 0 and k < n:

        y[i] += c[k] * weights[j]

  return y


def gaussianSmooth (y, sigma, truncate = 4.0):
  """
  Applies a Gaussian smoothing to the y array 
  """
  lw  = int (truncate * sigma + 0.5); m = 2 * lw + 1
  sum = 1.0; sigma2 = sigma ** 2; weights = np.empty (m)

  weights[lw] = 1.0

  for i in range (1, lw + 1):

    weight = np.exp (-0.5 * i ** 2 / sigma2)

    weights[lw + i] = weight;
    weights[lw - i] = weight;

    sum += 2 * weight;

  weights /= sum

  return convolve (y, weights)
```

```
def dirac (x, a):
  """
  Computes the Dirac delta
  shifted by a
  """
  dx = x[1] - x[0]

  i  = int (a / dx + 0.5) 
  y  = np.zeros (x.size)

  y[i] = 1.0 / dx

  return y
```

```
def boxCar (x, a, width):
  """
  Computes the Boxcar function
  shifted by a
  """
  n = x.size
  y = np.zeros (n)

  for i in range (n):

    if x[i] >= a - 0.5 * width and x[i] <= a + 0.5 * width:

      y[i] = 1
  
  return y
```

####2D-Gaussian smoothing

Gaussian smoothing works in any number of dimensions. In 2D, the Gaussian *kernel* takes the following form:
<br>
<br>
\begin{equation}
  G_{\sigma_x,\,\sigma_y}\,(x,\,y)
  \,=\,\frac{e^{-\frac{1}{2}\left(\frac{x^2}{{\sigma_x}^2}\,+\,\frac{y^2}{{\sigma_y}^2}\right)}}{2\pi\sigma_x\sigma_y}
\end{equation}
<br>
<br>

```
# Set the parameters below

# N: number of points

# xmin: left boundary of the domain
# xmax: right boundary of the domain
# ymin: bottom boundary of the domain
# ymax: top boundary of the domain

# sigma_x: half width of the Gaussian
#          in the x direction
# sigma_y: half width of the Gaussian
#          in the y direction

if __name__ == "__main__":

  N = 500

  xmin = -10
  xmax =  10
  ymin = -10
  ymax =  10

  sigma_x = 2
  sigma_y = 4
```

####3D-Gaussian smoothing

In 3D, the Gaussian *kernel* is nothing but a product of the three one-dimensional Gaussian funtions:
<br>
<br>
\begin{equation}
  G_{\sigma_x,\,\sigma_y,\,\sigma_z}\,(x,\,y,\,z)
  \,=\,\frac{e^{-\frac{1}{2}\left(\frac{x^2}{{\sigma_x}^2}
  \,+\,\frac{y^2}{{\sigma_y}^2}
  \,+\,\frac{z^2}{{\sigma_z}^2}\right)}}{N},
  \quad\text{where}\quad
  N
  \,=\,\int_{-\infty}^{\infty}\int_{-\infty}^{\infty}\int_{-\infty}^{\infty}
  e^{-\frac{1}{2}\left(\frac{x^2}{{\sigma_x}^2}
  \,+\,\frac{y^2}{{\sigma_y}^2}
  \,+\,\frac{z^2}{{\sigma_z}^2}\right)}\,dx\,dy\,dz
\end{equation}
<br>
<br>

####Smoothing in a GLL mesh

We can apply Gaussian smoothing to virtually any kind of grid, including a Gauss-Lobatto-Legendre (GLL) *mesh*.
<br>
<br>
For large data sets, we can dramatically reduce the computational effort by restricting the number of points used. We know that, for a distance larger than $3\sigma$, the weights become very small. Therefore, by using only samples within a $3\sigma$ radius, we can save many computations.
<br>
<br>
In the case of GLL *meshes*, it is also a good idea to further optimize the code by selecting points using the distances between the center of the *spectral elements* instead of comparing every single possible pair of samples.
<br>
<br>

#####Cartesian grid

```
def distance (x1, y1, x2, y2):
  """
  Cumputes the Cartesian distance in 2D
  """
  return np.sqrt ((x2 - x1) ** 2 + (y2 - y1) ** 2)
```

######Search condition

\begin{equation}
  d\,=\,\sqrt{{\Delta x}^2\,+\,{\Delta y}^2}
\end{equation}
<br>
<br>
\begin{equation}
  \frac{{\Delta x}^2}{{\left(3\sigma_x\right)}^2}
  \,+\,\frac{{\Delta y}^2}{{\left(3\sigma_y\right)}^2}
  \,\le\,1
  \,\implies\,
  \frac{{\Delta x}^2}{{\sigma_x}^2}
  \,+\,\frac{{\Delta y}^2}{{\sigma_y}^2}
  \,\le\,9
\end{equation}
<br>
<br>
Optimized version:
<br>
<br>
\begin{equation}
  \frac{{\Delta x_c}^2}{{\left(3\sigma_x\,+\,r_1\,+\,r_2\right)}^2}
  \,+\,\frac{{\Delta y_c}^2}{{\left(3\sigma_y\,+\,r_1\,+\,r_2\right)}^2}
  \,\le\,1
\end{equation}
<br>
<br>

#####Polar grid

For a polar grid, the horizontal distance $x$ becomes the arc distance $r\Delta\theta$ between two poins $A$ and $B$. The vertical distance $y$ is just the difference between the radii $r_1$ and $r_2$.

#####Spherical grid

In a spherical grid, we can calculate the distance between two points $A$ and $B$ with coordinates $(\theta_1,\,\phi_1)$ and $(\theta_2,\,\phi_2)$ using one of the formulae for arc distances:
<br>
<br>
Haversine formula
<br>
<br>
\begin{equation}
  \Delta \psi
  \,=\,2\arcsin{\left(\sqrt{\sin^2{\frac{\Delta \theta}{2}}
  \,+\,\cos{\phi_1}\cos{\phi_2}\sin^2{\frac{\Delta \phi}{2}}}\right)} 
\end{equation}
<br>
<br>
Vincenty formula
<br>
<br>
\begin{equation}
  \Delta \psi
  \,=\,\arctan{\left[\frac{\sqrt{\left(\cos{\theta_2}\sin{\Delta\phi}\right)^2
  \,+\,\left(\cos{\theta_1}\sin{\theta_2}
  \,-\,\sin{\theta_1}\cos{\theta_2}\cos{\Delta\phi}\right)^2}}{\sin{\theta_1}\sin{\theta_2}
  \,+\,\cos{\theta_1}\cos{\theta_2}\cos{\Delta\phi}}\right]} 
\end{equation}
<br>
<br>
Vector version
<br>
<br>
\begin{split}
  \Delta \psi
  \,=\,\arccos{\left(\mathbf{n}_1\,.\,\mathbf{n}_2\right)}
  \,&=\,\arccos{\left(\frac{v_1}{\left\lVert v_1\right\rVert}
  \,.\,\frac{v_2}{\left\lVert v_2\right\rVert}\right)}\\
  \,&=\,\arccos{\left[\frac{x_1x_2\,+\,y_1y_2\,+\,z_1z_2}{\sqrt{\left({x_1}^2\,+\,{y_1}^2\,+\,{z_1}^2\right)\left({x_2}^2\,+\,{y_2}^2\,+\,{z_2}^2\right)}}\right]}
\end{split}
<br>
<br> 
The vector version is a convenient choice for smoothing a *GLL* mesh as all the vectors corresponding to each mesh point $\mathbf{n}_1$ and $\mathbf{n}_2$ can be computed in advance, leading to faster code.
<br>
<br> 

```
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.patches import FancyArrowPatch, PathPatch
from mpl_toolkits.mplot3d import proj3d
```

```
def rThetaPhi2XYZ (r, t, p):
  """
  Converts spherical to Cartesian coordinates
  """
  x = r * np.sin (t) * np.cos (p);
  y = r * np.sin (t) * np.sin (p);
  z = r * np.cos (t);

  return x, y, z


def crossProduct (x1, y1, z1, x2, y2, z2):
  """
  Computes the cross-product between two 3D arrays
  """
  u, v, w = y1 * z2 - z1 * y2, z1 * x2 - x1 * z2, x1 * y2 - y1 * x2

  norm = np.sqrt (u ** 2 + v ** 2 + w ** 2)

  return u / norm, v / norm, w / norm


def vincenty (t1, p1, t2, p2):
  """
  Computes the great circle distance
  using the Vincenty's formula
  """
  sin_t1 = np.sin (t1)
  cos_t1 = np.cos (t1)
  sin_t2 = np.sin (t2)
  cos_t2 = np.cos (t2)

  sin_dp = np.sin (p2 - p1);
  cos_dp = np.cos (p2 - p1);

  return np.arctan2 (np.sqrt ((sin_t2 * sin_dp) ** 2
                              + (sin_t1 * cos_t2\
                              - cos_t1 * sin_t2 * cos_dp) ** 2),
                     cos_t1 * cos_t2 + sin_t1 * sin_t2 * cos_dp)


def greatCircle (r, t1, p1, t2, p2, N):
  """
  Creates a great circle
  """
  c = np.linspace (0, vincenty (t1, p1, t2, p2), N)

  x1, y1, z1 = rThetaPhi2XYZ (r, t1, p1)
  x2, y2, z2 = rThetaPhi2XYZ (r, t2, p2)

  u, v, w = crossProduct (x1, y1, z1, x2, y2, z2)

  x = np.empty (N)
  y = np.empty (N)
  z = np.empty (N)

  for i in range (len (c)):

    sinc = np.sin (c[i])
    cosc = np.cos (c[i])

    R = np.matrix ([[cosc + (u ** 2) * (1 - cosc),
                     u * v * (1 - cosc) - w * sinc,
                     u * w * (1 - cosc) + v * sinc],
                    [v * u * (1 - cosc) + w * sinc,
                     cosc + (v ** 2) * (1 - cosc),
                     v * w * (1 - cosc) - u * sinc],
                    [w * u * (1 - cosc) - v * sinc,
                     w * v * (1 - cosc) + u * sinc,
                     cosc + (w ** 2) * (1 - cosc)]])

    p = np.matrix ([[x1],
                    [y1],
                    [z1]])

    x[i], y[i], z[i] = R * p

  return x, y, z
```

```
if __name__ == '__main__':

  fig = plt.figure (figsize = (3.5, 3.5), dpi = 200)
  ax = fig.add_subplot (111, projection = '3d')

  r = 1.0

  t1, p1 = np.radians (35), np.radians (65)
  t2, p2 = np.radians (59), np.radians (48)

  x1, y1, z1 = rThetaPhi2XYZ (1.5 * r, t1, p1)
  x2, y2, z2 = rThetaPhi2XYZ (r, t2, p2)

  r1 = Arrow3D ([-0.02, x1], [-0.02, y1], [-0.02, z1],
                mutation_scale = 7,
                linewidth = 1.0,
                arrowstyle = '-|>',
                color = 'red',
                zorder = 1)
  r2 = Arrow3D ([-0.02, x2], [-0.02, y2], [-0.02, z2],
                mutation_scale = 7,
                linewidth = 1.0,
                arrowstyle = '-|>',
                color = 'red',
                zorder = 1)
    
  ax.add_artist (r1)
  ax.add_artist (r2)

  ax.text2D (0.528, 0.458,
             r'$r_1$',
             fontsize = 10,
             color = 'red',
             transform = ax.transAxes)
  ax.text2D (0.530, 0.700,
             r'$r_2$',
             fontsize = 10,
             color = 'red',
             transform = ax.transAxes)
  ax.text2D (0.6, 0.51,
             r'$A$',
             fontsize = 10,
             color = 'blue',
             transform = ax.transAxes)
  ax.text2D (0.71, 0.88,
             r'$B$',
             fontsize = 10,
             color = 'blue',
             transform = ax.transAxes)

  x, y, z = greatCircle (0.96 * r, t1, p1, t2, p2, 20)

  ax.plot (x, y, z,
           color = 'green',
           linestyle = 'dashed',
           linewidth = 0.7,
           zorder = 2)

  ax.text2D (0.620, 0.610,
             r'$r_1\Delta\psi$',
             fontsize = 10,
             color = 'green',
             transform = ax.transAxes)

  axis1 = Arrow3D ([-0.02, 1], [0, 0], [0, 0],
                   mutation_scale = 10,
                   lw = 1.5,
                   arrowstyle = '-|>',
                   color = 'black',
                   zorder = 0)
  axis2 = Arrow3D ([0, 0], [-0.02, 1], [0, 0],
                   mutation_scale = 10,
                   lw = 1.5,
                   arrowstyle = '-|>',
                   color = 'black',
                   zorder = 0)
  axis3 = Arrow3D ([0, 0], [0, 0], [-0.02, 1],
                   mutation_scale = 10,
                   lw = 1.5,
                   arrowstyle = '-|>',
                   color = 'black',
                   zorder = 0)

  ax.add_artist (axis1)
  ax.add_artist (axis2)
  ax.add_artist (axis3)

  ax.text2D (0.180, 0.230,
             r'$x$',
             fontsize = 10,
             transform = ax.transAxes)
  ax.text2D (0.880, 0.360,
             r'$y$',
             fontsize = 10,
             transform = ax.transAxes)
  ax.text2D (0.420, 0.910,
             r'$z$',
             fontsize = 10,
             transform = ax.transAxes)

  ax.set_axis_off ()

  limits = np.array([getattr (ax, f'get_{axis}lim') () for axis in 'xyz'])
  ax.set_box_aspect (np.ptp (limits, axis = 1))

  ax.set_xlim (-0.2, 0.9)
  ax.set_ylim (-0.2, 0.9)
  ax.set_zlim (-0.2, 0.9)

  ax.view_init (elev = 25, azim = 25)

  plt.subplots_adjust (left = 0.0, bottom = 0.0, top = 1.0, right = 1.0)

  plt.show ()
```
