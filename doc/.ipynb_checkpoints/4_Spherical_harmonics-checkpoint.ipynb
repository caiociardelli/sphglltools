####Trigonometric parametrization of the associated Legendre polynomials

It is useful to parametrize the *associated Legendre polynomials* in terms of the $\cos\,(\theta)$. Letting $\,x\,=\,\cos\,(\theta)\,\,\text{over}\,\,\left[0,\,\pi\right]$:
<br>
<br>
\begin{equation}
  P_{n}^{m}\,(\cos\theta)
  \,=\,\frac{(-1)^m}{2^n n!}\,\left(\sin\theta\right)^m\,\left[\frac{d}{d(\cos\theta)}\right]^{n\,+\,m}\left(\cos^2\theta\,-\,1\right)^n\qquad 0\,\le\,m\,\le\,n\quad m,\,n \in \mathbb{N}
\end{equation}
<br>
<br>
These functions satisfy the fixed-order orthonormality relation:
<br>
<br>
\begin{split}
  \int_{0}^{\pi} P_{n}^{m}\,(\cos\theta)\,P_{n'}^{m}\,(\cos\theta)\,\sin\theta\,d\theta
  \,=\,
  \begin{cases}
    \displaystyle 1 \quad \text{if}\quad n\,\ne\,n'\\\\
    \displaystyle \frac{2}{2n\,+\,1}\frac{(n\,+\,m)!}{(n\,-\,m)!}\quad \text{otherwise}
  \end{cases}
\end{split}
<br>
<br>
When $m\,=\,0$ the functions reduce to the ordinary *Legendre polynomials* given by the *Rodrigues' formula*:
<br>
<br>
\begin{equation}
  P_n\,(\cos\,\theta)
  \,=\,\,\frac{1}{2^n n!}\,\left[\frac{d}{d(\cos\theta)}\right]^{n}\left(\cos^2\theta\,-\,1\right)^n
\end{equation}
<br>
<br>

####Spherical harmonics

We can combine the *associated Legendre polynomials* with the trigonometric functions $\sin m\phi$ and $\cos m\phi$ to create a complete two-dimensional set of basis functions, useful for expanding 2D surfaces. For surfaces with spherical symmetry, these basis functions are known as *spherical harmonics*.
<br>
<br>
As described by *Dahlen & Tromp (1998)*, *spherical harmonics* can be real or complex functions, depending on the application. For *quantum mechanics*, the complex basis functions provide a more convenient way to represent the wave functions. On the other hand, geophysicists usually deal with real fields, so the real spherical harmonics seem more appropriate.
<br>
<br>
Following the convention of *Dahlen & Tromp (1998)*, we shall use a normalized version $X_{n,\,m}\,(\cos\,\theta)$ of the *associated Legendre polynomials* $P_{n}^{m}\,(\cos\,\theta)$ in the following form:
<br>
<br>
\begin{equation}
  X_{n,\,m}\,(\cos\,\theta)
  \,=\,\sqrt{(2\,-\,\delta_{m}^{0})\frac{(2n\,+\,1)}{4\pi}\frac{(n\,-\,m)!}{(n\,+\,m)!}}\,P_{n}^{m}\,(\cos\theta),
\end{equation}
<br>
where $\delta_{m}^{0}$ is the Kronecker delta:
<br>
<br>
\begin{equation}
  \delta_{m}^{0}
  \,=\,
  \begin{cases}
    1\quad\text{if}\quad m\,=\,0\\
    0\quad\text{otherwise}
  \end{cases}
\end{equation}
<br>
<br>
Therefore, we can expand a spherically symmetric function $f\,(\theta,\phi)$ into *spherical harmonics* in the following way:
<br>
<br>
\begin{equation}
  f\,(\theta,\phi)
  \,=\,
  \sum_{n\,=\,0}^{\infty}
  \sum_{m\,=\,0}^{n}
  \left(a_{m,\,n}\cos m\phi
  \,+\,
  b_{m,\,n}\sin m\phi\right)\,X_{n,\,m}\,(\cos\theta)
\end{equation}
<br>
<br>
\begin{equation}
  a_{m,\,n}
  \,=\,
  \int_{-\pi}^{\pi}
  \int_{0}^{\pi}
  f\,(\theta,\phi)\,X_{n,\,m}\,(\cos\theta)\,\cos m\phi\,\sin\theta\,d\theta\,d\phi\\
  b_{m,\,n}
  \,=\,
  \int_{-\pi}^{\pi}
  \int_{0}^{\pi}
  f\,(\theta,\phi)\,X_{n,\,m}\,(\cos\theta)\,\sin m\phi\,\sin\theta\,d\theta\,d\phi
\end{equation}
<br>
<br>

```
import numpy as np

from scipy.special import lpmv, factorial, factorial2
from scipy.integrate import dblquad
```

```
def lgP (m, n, x):
  """
  Evaluates the normalized P_{m}^{n}(x) for a given m, n
  (m <= n <= nmax) and x (-1 <= x <= 1)
  """
  k1 = 1 if m == 0 else 2
  k2 = (2 * n + 1) / (4 * np.pi)
  k3 = factorial (n - m) / factorial (n + m)

  nml = np.sqrt (k1 * k2 * k3)

  return nml * lpmv (m, n, x)


def coefficients (nmax, f):
  """
  Evaluates a_{m,n} and b_{m,n}
  """
  a = np.zeros ((nmax + 1, nmax + 1))
  b = np.zeros ((nmax + 1, nmax + 1))

  for n in range (nmax + 1):

    for m in range (n + 1):

      fa = lambda theta, phi: f (theta, phi)\
                              * lgP (m, n, np.cos (theta))\
                              * np.cos (m * phi)\
                              * np.sin (theta)

      fb = lambda theta, phi: f (theta, phi)\
                              * lgP (m, n, np.cos (theta))\
                              * np.sin (m * phi)\
                              * np.sin (theta)

      a[m, n] = dblquad (fa, -np.pi, np.pi,
                         lambda theta: 0,
                         lambda theta: np.pi)[0]

      b[m, n] = dblquad (fb, -np.pi, np.pi,
                         lambda theta: 0,
                         lambda theta: np.pi)[0]

      if abs (a[m, n]) > 1e-5:

        print ('a^{}_{} = {:+E}'.format (m, n, a[m, n]))  

      if abs (b[m, n]) > 1e-5:

        print ('b^{}_{} = {:+E}'.format (m, n, b[m, n]))

  return a, b


def sCS (N, nmax, a, b, theta, phi):
  """
  Evaluates S(theta, phi)
  """
  F = np.zeros ((N, 2 * N))

  PHI, THETA = np.meshgrid (phi, theta)

  for n in range (nmax + 1):

    for m in range (nmax + 1):

      Pmn = lgP (m, n, np.cos (THETA))

      Cmn = Pmn * np.cos (m * phi)
      Smn = Pmn * np.sin (m * phi)

      F += a[m, n] * Cmn + b[m, n] * Smn
  
  return F
```

```
# Set the parameters below

# N: number of points

# nmax: maximum degree

if __name__ == "__main__":

  N = 200
  
  nmax = 9
```

```
# Uncomment one of the following functions or define your own

if __name__ == "__main__":

  f = lambda theta, phi: 1 + 0.1\
                         * np.sin (6 * theta)\
                         * np.sin (6 * phi)

  # f = lambda theta, phi: 1 + 0.15\
  #                        * np.sin (8 * theta)\
  #                        * np.cos (4 * phi)

  # f = lambda theta, phi: 1 + 0.2\
  #                        * np.cos (3 * theta)\
  #                        * np.sin (7 * phi)
```

```
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm, colors
```

```
if __name__ == "__main__":

  theta = np.linspace (0, np.pi, N)
  phi   = np.linspace (-np.pi, np.pi, 2 * N)

  PHI, THETA = np.meshgrid (phi, theta)

  fig, ax = plt.subplots (subplot_kw = dict (projection = '3d'),
                          figsize = (5, 4), dpi = 140)

  F = f (THETA, PHI)

  X = F * np.sin (THETA) * np.cos (PHI)
  Y = F * np.sin (THETA) * np.sin (PHI)
  Z = F * np.cos (THETA)

  norm = colors.Normalize ()

  ax.plot_surface (X, Y, Z, rstride = 1, cstride = 1,
                   facecolors = cm.jet_r (norm (F)))

  ax.set_title (r'$f\,(\theta,\,\phi)$', fontsize = 16)
  m = cm.ScalarMappable (cmap = cm.jet_r)
  m.set_array (F)
  fig.colorbar (m,
                aspect = 20,
                ax = plt.gca ())
  
  ax.set_box_aspect ((np.ptp (X),
                      np.ptp (Y),
                      np.ptp (Z)))

  ax.set_xlim (-1, 1)
  ax.set_ylim (-1, 1)
  ax.set_zlim (-1, 1)
  ax.axis ('off')

  plt.show ()

  fig, ax = plt.subplots (subplot_kw = dict (projection = '3d'),
                          figsize = (5, 4), dpi = 140)

  a, b = coefficients (nmax, f)

  S = sCS (N, nmax, a, b, theta, phi)

  X = S * np.sin (THETA) * np.cos (PHI)
  Y = S * np.sin (THETA) * np.sin (PHI)
  Z = S * np.cos (THETA)
  
  ax.plot_surface (X, Y, Z, rstride = 1, cstride = 1,
                   facecolors = cm.jet_r (norm (S)))
  
  ax.set_title (r'$S_{{{}}}\,(\theta,\,\phi)$'.format (nmax),
                fontsize = 16)

  fig.colorbar (m,
                aspect = 20,
                ax = plt.gca ())

  ax.set_box_aspect ((np.ptp (X),
                      np.ptp (Y),
                      np.ptp (Z)))

  ax.set_xlim (-1, 1)
  ax.set_ylim (-1, 1)
  ax.set_zlim (-1, 1)
  ax.axis ('off')

  plt.show ()
```
