####Integral identities

The following integral identities are very useful for the derivation of the trigonometric series:
<br>
<br>
\begin{split}
  &\int_{-\pi}^{\pi} \sin\,(mx)\,dx
  \,=\,
  0\qquad\forall\,\,\,m\,\in\,\mathbb{N}^*\\\\
  &\int_{-\pi}^{\pi} \cos\,(mx)\,dx
  \,=\,
  0\qquad\forall\,\,\,m\,\in\,\mathbb{N}^*\\\\
    &\int_{-\pi}^{\pi} \sin\,(mx)\cos\,(nx)\,dx
  \,=\,
  0\qquad\forall\,\,\,m,n\,\in\,\mathbb{N}^*\\\\
  &\int_{-\pi}^{\pi} \sin\,(mx)\sin\,(nx)\,dx
  \,=\,
  \begin{cases}
    0\qquad\text{if}\qquad m\,\ne\,n\\
    \pi\qquad\text{if}\qquad m\,=\,n
  \end{cases}\\\\
  &\int_{-\pi}^{\pi} \cos\,(mx)\cos\,(nx)\,dx
  \,=\,
  \begin{cases}
    0\qquad\text{if}\qquad m\,\ne\,n\\
    \pi\qquad\text{if}\qquad m\,=\,n
  \end{cases}
\end{split}
<br>
<br>

####Trigonometric series

The idea is trying to expand a periodic function $f\,(x)$ into a series in the following form:
<br>
<br>
\begin{equation}
  f\,(x)
  \,=\,a_0
  \,+\,a_1\cos\,(x)\,+\,b_1\sin\,(x)
  \,+\,a_2\cos\,(2x)\,+\,b_2\sin\,(2x)
  \,+\,...
  \,+\,a_m\cos\,(mx)\,+\,b_n\sin\,(nx)
\end{equation}
<br>
<br>
To determine the coefficients, we can use the integral identities for the trigonometric functions. For the $a_0$ coefficient, we just need to integrate both members of the equation above over the interval $[-\pi,\,\pi]$:
<br>
<br>
\begin{equation}
  \int_{-\pi}^{\pi} f\,(x)\,dx
  \,=\,\int_{-\pi}^{\pi} \left[a_0
  \,+\,a_1\cos\,(x)\,+\,b_1\sin\,(x)
  \,+\,a_2\cos\,(2x)\,+\,b_2\sin\,(2x)
  \,+\,...
  \,+\,a_n\cos\,(nx)\,+\,b_n\sin\,(nx)\right]\,dx
\end{equation}
<br>
<br>
The terms involving all the coefficients but $a_0$ vanish and what is left is:
<br>
<br>
\begin{equation}
  \int_{-\pi}^{\pi} f\,(x)\,dx
  \,=\,
  2\pi\,a_0
  \,\implies\,
  a_0
  \,=\,
  \frac{1}{2\pi} \int_{-\pi}^{\pi} f\,(x)\,dx
\end{equation}
<br>
<br>
To calculate the $a_m$ coefficients we multiply both members by $\cos\,(nx)$ and integrate over $[-\pi,\,\pi]$, eliminating all terms but $a_n\,\cos\,(nx)$:
<br>
<br>
\begin{equation}
  \int_{-\pi}^{\pi} f\,(x)\,\cos\,(nx)\,dx
  \,=\,
  \pi\,a_n\,\cos\,(nx)
  \,\implies\,
  a_n
  \,=\,
  \frac{1}{\pi} \int_{-\pi}^{\pi} f\,(x)\,\cos\,(nx)\,dx
\end{equation}
<br>
<br>
Similarly, we can extract all the $b_n$ coefficients by multipliying both members by $\sin\,(nx)$ and again integrating over $[-\pi,\,\pi]$:
<br>
<br>
\begin{equation}
  \int_{-\pi}^{\pi} f\,(x)\,\sin\,(nx)\,dx
  \,=\,
  \pi\,b_n\,\sin\,(nx)
  \,\implies\,
  b_n
  \,=\,
  \frac{1}{\pi} \int_{-\pi}^{\pi} f\,(x)\,\sin\,(nx)\,dx
\end{equation}
<br>
<br>
Therefore, the *Fourier series* representation of a periodic function $f\,(x)$ is:
<br>
<br>
\begin{equation}
  f\,(x)
  \,=\,
  \frac{a_0}{2}
  \,+\,
  \sum_{n\,=\,1}^{\infty}\left[\,a_n\,\cos\,(nx)
  \,+\,
  \,b_n\,\sin\,(nx)\right],
\end{equation}
<br>
<br>
where
<br>
<br>
\begin{split}
  &a_0
  \,=\,
  \frac{1}{\pi}
  \int_{-\pi}^{\pi} f\,(x)\,dx\\\\
  &a_n
  \,=\,
  \frac{1}{\pi}
  \int_{-\pi}^{\pi} f\,(x)\,\cos\,(nx)\,dx\\\\
  &b_n
  \,=\,
  \frac{1}{\pi}
  \int_{-\pi}^{\pi} f\,(x)\,\sin\,(nx)\,dx
\end{split}
<br>
<br>

```python
import numpy as np
import matplotlib.pyplot as plt

from numpy import  vectorize
from scipy.integrate import simps
```

```python
# Set the parameters below

# N: number of points

# nmax: maximum degree

if __name__ == "__main__":

  N = 1000

  nmax = 6
```

```python
def periodic (x):
  """
  Auxiliary function to create periodic functions
  """
  return x % (2 * np.pi)
```

```python
# Uncomment one of the following functions or define your own

if __name__ == "__main__":

  def f (x):

    return 3 * periodic (x) + 10

   @vectorize
  # def f(x):

  #   if 0.0 <= periodic (x) <= np.pi: return 2.0

  #   else: return 1.0
  
  # def f (x):

  #   x = periodic (x)

  #   conditions = [x < 0, (x > 0) & (x < 1), (x > 1) & (x < 2), x > 2]
  #   functions  = [lambda x: x + 6.0, lambda x: 6.0, 
  #                 lambda x: 6.0 - x, lambda x: (3.5 - x) ** 2 + 6.0]

  #   return np.piecewise (x, conditions, functions)
```

```python
def coefficients (nmax, f):
  """
  Evaluates a_{0}, a_{n} and b_{n}
  """
  x = np.linspace (-np.pi, np.pi, N)

  a = np.zeros (nmax + 1)
  b = np.zeros (nmax + 1)

  a[0] = simps (f (x), x) / (2 * np.pi) 

  for n in range (1, nmax + 1):

    a[n] = simps (f (x) * np.cos (n * x), x) / np.pi
    b[n] = simps (f (x) * np.sin (n * x), x) / np.pi

  return a, b


def sCS (nmax, a, b, x):
  """
  Evaluates S(x)
  """
  s = np.zeros (x.size)

  for n in range (nmax + 1):

    s += a[n] * np.cos (n * x) + b[n] * np.sin (n * x)
  
  return s
```

```python
if __name__ == "__main__":

  x = np.linspace (-5 * np.pi, 5 * np.pi, N)

  a, b = coefficients (nmax, f)
  s = sCS (nmax, a, b, x)

  fig = plt.figure (figsize = (6, 4), dpi = 140)

  for n in range (nmax + 1):

    plt.plot (x, a[n] * np.cos (n * x), linewidth = 1.2)
    plt.plot (x, b[n] * np.sin (n * x), linewidth = 1.2)

  plt.plot (x, f (x), color = 'red', linewidth = 1.2,
            label = r'$f\,(x)$')
  plt.plot (x, s, color = 'black', linestyle = 'dashed', linewidth = 1.2,
            label = r'$S\,(x)$')
  plt.xlabel (r'$x$', fontsize = 14)
  plt.ylabel (r'$y$', fontsize = 14)
  plt.tick_params (axis = 'x', labelsize = 10)
  plt.tick_params (axis = 'y', labelsize = 10)
  fig.legend (loc = 'center right', shadow = True, framealpha = 1.0,
              handlelength = 1.8, fontsize = 10)
  plt.grid (linestyle = 'dashed', linewidth = 0.5)
  plt.title ('Fourier expansion', fontsize = 16, y = 1.03)

  plt.subplots_adjust (right = 0.83)

  plt.show ()
```
